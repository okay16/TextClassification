```python
from __future__ import absolute_import, division, print_function
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
imdb = keras.datasets.imdb
```

```python
#splitting the dataset into training and testing
```

```python
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
print(f"Training entries {len(train_data)}. Labels: {len(train_labels)}")
```

```python
print(train_data[0]) #each number is mapped to a word in a dict
```

```python
print(len(train_data[0])), print(len(train_data[1])) #number of words in first and second review
```

```python
word_index = imdb.get_word_index()
```

```python
word_index = {k: (v+3) for k, v in word_index.items()}
```

```python
word_index["<PAD>"] = 0
```

```python
word_index["<START>"] = 1
```

```python
word_index["<UNK>"] = 2
```

```python
word_index["<UNUSED>"] = 3
```

```python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()]) #flipping value key to key value
```

```python
#takes in a list of integers and returns an english version of it
def decode_review(text):
    return ' '.join([reverse_word_index.get(i, "?") for i in text])
```

```python
print(decode_review(train_data[0]))
```

```python
print(train_data[0])
```

```python
train_data = keras.preprocessing.sequence.pad_sequences(train_data,
                                                        value=word_index["<PAD>"],
                                                        padding="post",
                                                        maxlen=256)
```

```python
test_data = keras.preprocessing.sequence.pad_sequences(test_data,
                                                        value=word_index["<PAD>"],
                                                        padding="post",
                                                        maxlen=256)
```

```python
print(len(train_data[0]), len(train_data[1]))
```

```python
#constructing the Neural Network using Tensorflow Keras
#optimised all reviews to be of equal length 
```

```python
#how many layers? How many hidden units per model?
```

```python
vocab_size = 10000
```

```python
#adding layers + their description
model = keras.Sequential()
```

```python
#input layer

model.add(keras.layers.Embedding(vocab_size, 16))
```

```python
#hidden layer 1 returns a fixed output vector for each review by averaging over the sequence dimension
#allows model to handle input length
model.add(keras.layers.GlobalAveragePooling1D())
```

```python
#hidden layer 2
#16 hidden units
model.add(keras.layers.Dense(16, activation=tf.nn.relu))
```

```python
#output layer will either indictate 0 or 1 = output of the neural network
#does this using the sigmoid ativation function = floating point value = confidence level of the NN to be either positive or negative

model.add(keras.layers.Dense(1, activation=tf.nn.sigmoid))
```

```python
model.build()
```

```python
#gives a summary of the NN
#model.summary()
```

```python
# 0 = negative review
# 1 = positive review
# => Binary Classification
```

```python
# making a loss function 
```

```python
model.compile(optimizer="adam",
             loss="binary_crossentropy",
             metrics=["acc"])
```
